```python
# Task_1: Safe Division Utility

try:
    numerator = int(input("Enter the numerator value"))
    denominator = int(input("Enter the denominator value"))
    result = numerator/denominator
    
except ValueError as e:
    print("Please provide the integer data",e)
except ZeroDivisionError as ex:
    print(ex)
else:
    print("The Result is :",result)
finally:
    print("Division Operation completed")
```

```python
# Task -2 Bill Calculator_with Error_Handling
prices = [120,350,'abc',500,-200,800]
total = 0
for i in prices:
    try:
        i= int(i)
        if i<0:
            raise ValueError("Negative price not allowed")
        total = total + i
        print("Running total",total)
    except ValueError as e:
        print(e)
            
```

```python
#Task -3: Custom Exception Age Validator
def check_age(age):
    if age < 1 or age > 120:
        raise ValueError("Age must be between 1 and 120")
    else:
        print("Age is Valid")
try:
    age = int(input("Provide your age"))
    check_age(age)
except ValueError as e:
    print("Error",e)
```

```python
#Task -4 File Reader with Exception Handling
try:
    file_name = input("enter fiulename")
    file = open(file_name,'r')
    for i in range(3):
        line = file.readline()
        if not line:
            break
        print(line.strip())
    file.close()
except FileNotFoundError as e:
    print("Error",e)
except PermissionError as e1:
    print("Error",e1)
finally:
    print("File operation attempted")
```

```python
#Task -5 Mini Program: Safe Shopping Cart
cart = []
while True:
    user_input = input("Enter Price")
    if user_input.lower() == 'q':
        break
    try:
        price = float(user_input)
        if price <0:
            raise ValueError("price is Negative")
        cart.append(price)
    except ValueError as e:
        print("Invalid Input",e)
print("Total items",len(cart))
print("Total Bill",sum(cart))
```
