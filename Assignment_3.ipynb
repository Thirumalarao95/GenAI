```python
#Task-1
def apply_discount(price,discount_percent = 5):
    if discount_percent <= 60:
        discount_amount = price*(discount_percent/100)
        price_after_discount = price - discount_amount
        return price_after_discount
print(f'The Price after discount is :{apply_discount(1000,10)}')
print(f'The Price after discount is :{apply_discount(500)}')
print(f'The Price after discount is :{apply_discount(2000,70)}')
```

```python
#Task -2
def factorial(n):
    factorial_value = 1
    if n == 0 or n==1:
        return factorial_value
    elif n<0:
        print('n is Negative Number')
    else:
        return n * factorial(n-1)
print(factorial(5))
print(factorial(0))
print(factorial(-3))
```

```python
#task-6
#prices = [100,500,900,500,750]
def process_prices(prices):
    discount_prices = list(map(lambda x: x - (x * 0.10),prices))
    filtered_prices = list(filter(lambda x: x>300,discount_prices))
    return discount_prices,filtered_prices
process_prices([100,500,900,500,750])
```

```python
#Task -7 
prices = []
#[100,250,400,1200,50,2000,850]
def add_price(prices_list,price):
    prices_list.append(price)
    print("The Price List is", prices_list)


def get_average_price(prices_list):
    total = 0
    if len(prices_list) ==0:
        print("Lstis empty")
    else:
        for i in prices_list:
            total = total + i
            length = len(prices_list)
            average = total/length
    print("Average Value is ",average)
        

def get_max_price(prices_list):
    if len(prices_list) >0:
        maximum = prices_list[0]
        for i in prices_list:
            if i > maximum:
                maximum = i
        print("Maximum Value is",maximum)
    else:
        print("List is empty")
        

while True:
    print("1. Add price")
    print("2. Show Average Price")
    print("3. Show maximum Price")
    print("q. Quit")
    ch = input("Enter your option")
    if ch == '1':
        price = int(input("please Enter price:"))
        add_price(prices,price)
    elif ch == '2':
        get_average_price(prices)
    elif ch == '3':
        get_max_price(prices)
    elif ch == 'q':
        break
    else:
        print("Please choose th correct option")
```
