```python
#Task 4:
#I opened the file in read mode and converted the data into int with the help of strip
f1 = open('sales_data.txt','r')
all_data = f1.readlines()
f1.close()
int_list = [int(line.strip()) for line in all_data]
length = len(int_list)
total = 0
#here i am calculating the total
for i in int_list:
    total = total +i

#here i am finding thw Highest sales
Highest_sale = int_list[0]
for i in int_list:
    if i> Highest_sale:
        Highest_sale = i
#here i finding the lowest sale
lowest_sale = int_list[0]
for i in int_list:
    if i < lowest_sale:
        lowest_sale = i
#Here I am calulating the average sales
average_sales = total/length
print(f"Total sales : {total}")
print(f"Highest sales : {Highest_sale}")
print(f"lowest sales : {lowest_sale}")
print(f"average sales : {average_sales}")
```

```python
#Task 5:
#I am asking the user for the 3 products and prices
products = []
for i in range(3):
    product_name = input("Enter product name")
    price = int(input("enter price of product"))
    products.append((product_name,price))
#I am write the data into products.txt file 
#now i am opening the file with write mode
f1 = open('products.txt','w')
for i in products:
    f1.write(i[0] + " | " +str(i[1]) + "\n")
f1.close()
f1 = open('products.txt','r')
for i in f1:
    product_name,price = i.strip().split(" | ")
    print(f"Product Name : {product_name}, Price: {price}")
f1.close()
```

```python
#Task-7
prices = {'Mouse':500,
          'Keyboard':800,
          'Monitor':7000,
          'Pendrive':400,
          'Camera':5000
         }
discount_percentage = float(input("Enter Discount Percentage"))
f1 = open('discount_report.txt','w')
f1.write('Product | Original Price | Discounted Price \n')
total_discounted = 0
count = 0
for i in prices:
    original = prices[i]
    discounted = original - (original * discount_percentage/100)
    f1.write(f"{i} | {original} | {discounted} \n")
    total_discounted += discounted
    count += 1
    average = total_discounted / count
    f1.write("\n ---- Summary ----")
    f1.write(f"Total Items: {count} \n")
    f1.write(f"Average Discounted Price: {average}\n")
f1.close()
f1 = open("discount_report.txt",'r')
print("\n Discount Report")
print(f1.read())
f1.close()
```
